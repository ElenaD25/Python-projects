```python
df.drop(columns = ['AREA','Mocodes','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4','Cross Street','Rpt Dist No','Part 1-2'], inplace = True)
```

```python
df.rename(columns={'DR_NO':'ID','Date Rptd':'Reported Date','DATE OCC':'Occured Date',
'TIME OCC':'Occured Time','AREA NAME':'Area Name','Crm Cd':'Crime Code','Crm Cd Desc':'Crime Code Desc',
'Vict Age':'Victim Age','Vict Sex':'Victim Sex','Vict Descent':'Race',
'Premis Cd':'Crime Loc Code','Premis Desc':'Crime Loc Code Desc','Weapon Used Cd':'Weapon Code',
'Weapon Desc':'Weapon Desc','LOCATION':'Location','LAT':'Latitude','LON':'Longitude'}, inplace = True)
```

```python
df.isna().sum()
```

```python
df['Victim Sex'] = df['Victim Sex'].fillna('Unknown')
df['Race'] = df['Race'].fillna('Unknown')
df['Crime Loc Code'] = df['Crime Loc Code'].fillna(0)
df['Crime Loc Code Desc'] = df['Crime Loc Code Desc'].fillna('Not specified')
df['Weapon Code'] =df['Weapon Code'].fillna(0)
df['Weapon Desc'] =df['Weapon Desc'].fillna('Weapon Desc')
```

```python
df['Reported Date'] = pd.to_datetime(df['Reported Date']).dt.date
df['Occured Date'] = pd.to_datetime(df['Occured Date']).dt.date
```

```python
df.loc[df['Occured Time'] == 1, 'Occured Time'] = 100
df.loc[df['Occured Time'] == 5, 'Occured Time'] = 500
df.loc[df['Occured Time'] == 2, 'Occured Time'] = 200
df.loc[df['Occured Time'] == 3, 'Occured Time'] = 300
df.loc[df['Occured Time'] == 4, 'Occured Time'] = 400
df.loc[df['Occured Time'] == 8, 'Occured Time'] = 800
df.loc[df['Occured Time'] == 9, 'Occured Time'] = 900
df.loc[df['Occured Time'] == 6, 'Occured Time'] = 600
df.loc[df['Occured Time'] == 7, 'Occured Time'] = 700

df['Occured Time'] = pd.to_datetime(df['Occured Time'], format='%H%M').dt.strftime('%H:%M') 
```

```python
df
```

```python
df['Weapon Code'] = df['Weapon Code'].astype(int)
df['Crime Loc Code'] = df['Crime Loc Code'].astype(int)
```

```python
df['Location'] = df['Location'].str.strip()
```

```python
def replace_race_code(x):
    if x == 'A':
        return 'Other Asian'
    elif x == 'B':
        return 'Black'
    elif x == 'C':
        return 'Chinese'
    elif x == 'D':
        return 'Cambodian'
    elif x == 'F':
        return 'Filipino'
    elif x == 'G':
        return 'Guamanian'
    elif x == 'H':
        return 'Hispanic'
    elif x == 'I':
        return 'American Indian'
    elif x == 'J':
        return 'Japanese'
    elif x == 'K':
        return 'Korean'
    elif x == 'L':
        return 'Laotian'
    elif x == 'O':
        return 'Other'
    elif x == 'P':
        return 'Pacific Islander'
    elif x == 'S':
        return 'Samoan'
    elif x == 'U':
        return 'Hawaiian'
    elif x == 'V':
        return 'Vietnamese'
    elif x == 'X':
        return 'Unknown'
    elif x == 'Z':
        return 'Asian Indian'
    
df['Race'] = df['Race'].apply(replace_race_code)

df
```

```python
df['Victim Sex'].value_counts()
```

```python
def replace_victim_sex_code(x):
    if x == 'F':
        return 'Female'
    elif x == 'M':
        return 'Male'
    else:
        return 'Unknown'
    
df['Victim Sex']= df['Victim Sex'].apply(replace_victim_sex_code)
df
    
```

```python
df['Crime Code Desc'] = df['Crime Code Desc'].str.title()
df['Crime Loc Code Desc'] = df['Crime Loc Code Desc'].str.title()
df['Location'] = df['Location'].str.title()
df['Weapon Desc']=df['Weapon Desc'].str.title()

df['Weapon Code'] = df['Weapon Code'].astype(int)
```
